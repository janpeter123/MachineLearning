# Random Forest Regression

## Importing the libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```
data = pd.read_csv('./Position_Salaries.csv')
```

```
data.head(3)
```

```
x = data.Level.values
y = data.Salary.values

x = x.reshape(len(x),1)
y = y.reshape(len(y),1)

print(x,'\n\n',y)
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor

#O argumento mais importante é o n_estimators, que é o número de arvores
regressor = RandomForestRegressor(random_state=0,n_estimators=10)
regressor.fit(x,y)
```

## Predicting a new result

```
regressor.predict([[2.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
x_hr = np.arange(min(x),max(x),0.01)
x_hr = x_hr.reshape(len(x_hr),1)
plt.scatter(x,y, color='red')
plt.plot(x_hr,regressor.predict(x_hr),color='blue')
```
